```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv("C:/Users/Gabo0/Desktop/Jupyter Python/Ejercicios Linkedin/base_datos_2008.csv", nrows = 1e4)
newdf= df[["AirTime","DepDelay"]].dropna()
```

```python
newdf = df[["AirTime","Distance","TaxiOut","ArrDelay","DepDelay"]].dropna()
```

```python
clstr = AgglomerativeClustering(n_clusters =5)
clstr.fit(newdf)

#Euclidean /L1 /L2 /manhattan / cosine / precomputed
# affinity = "l1", linkage="complete"
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c=clstr.fit_predict(newdf))
plt.show
```
