```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Gabo0/Desktop/Jupyter Python/Ejercicios Linkedin/base_datos_2008.csv")
```

Limpiamos la base, borramos los datos no encontrados y creamos un set con las con las que queremos trabajar.

```python
df.dropna(inplace = True, subset = ["ArrDelay","DepDelay", "Distance"])
```

Gráfico de distribución, este es un sistema clasico con la aproximación de la densidad

```python
sns.distplot(df["Distance"])
```

 Sin la aproximación añadimos el siguiente parametro y se pueden modificar el número de bins en el gráfico.

```python
sns.distplot(df["Distance"], kde = False, bins = 100)
```

Para comparar dos distribuciones =

```python
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-300,300])
```

Para ver como se realizan diagramas de caja más sencillos, se crea un nueo dataframe

```python
df2 = df[df["Origin"].isin(["ATL","HOU","IND"])].sample(frac=1).head(500)
```

```python
sns.boxplot(x="DepDelay",y="Origin", data =df2)
plt.xlim([-20,150])
```

La linea dentro de las cajas, representa la mediana, la primera linea de izqueirda a derecha de la caja representa el primer cuartil, la linea derecha de la caja representa el tercer cuartil. Los rombos representan los outliers, en el caso de este gráficos se identifican como los que exceden 1.5 veces el rango intercuartilico.
