```python
from pyspark import SparkConf, SparkContext
from pyspark.sql.types import StringType
from pyspark import SQLContext

conf = SparkConf().setMaster("local").setAppName("Mi programa")
sc = SparkContext(conf=conf)
sqlContext = SQLContext(sc)
```

```python
dfspark = sqlContext.read.format("csv").option("header", "true").option("inferSchema","true").load("C:/Users/Gabo0/Desktop/Jupyter Python/Ejercicios Linkedin/just tacos and burritos.csv")
```

```python
dfspark = dfspark.na.drop(subset=['cuisines'])
```

```python
dfspark.select("cuisines").take(50)
```

```python
A = sc.parallelize(dfspark.select("cuisines").rdd.collect())
```

```python
def split_cuisines(x):
    x = x[0]
    x = x.replace(" ","")
    x = x.split(",")
    return x
```

```python
flatmap = A.flatMap(split_cuisines)
```

```python
flatmap.collect()
```

```python
mapfun = flatmap.map(lambda x: (x,1))
```

```python
results = mapfun.reduceByKey(lambda x,y: x+y)
```

```python
results.sortBy(lambda x:x[1], ascending=False).collect()
```
