```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Gabo0/Desktop/Jupyter Python/Ejercicios Linkedin/base_datos_2008.csv", nrows = 100000)
data = np.unique(df.DayOfWeek, return_counts=True)
labs = ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"]
data
```

```python
plt.pie(x=data[1],
       #labels=labs,
       radius = 0.7,
       colors = sns.color_palette("hls",7),
        # explode separa las fracciones del diagrama circular
        explode = (0,0,0,0,0,0,0.1),
       startangle = 90,
        #autopct agregar porcentajes al diagrama circular
       autopct ="%1.1f%%",
       labeldistance = 1.2)
plt.legend(loc="right", labels = labs)
        
plt.show()
```

```python
plt = sns.barplot(x = labs, y = data[1])
plt.set(xlabel="Día de la semana", ylabel = "Número de vueelos")
```
